# Condicional

```python
# If
if 5 < 10:
  print('Verdade')
```

```python
# If/else
num1 = 10
num2 = 5
if num1 > num2:
  print('O num1 é o maior número')
else:
  print('O num2 é o maior número')
```

```python
# If / Elif / Else
num1 = 10
num2 = 10
if num1 > num2:
  print('O num1 é o maior número')
elif num1 < num2:
  print('Os num2 é o maior número')
else:
  print('Os dois números são igual')
```

# Condicional com operadores lógicos

```python
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print('O num1 é o maior número')
else:
  print('O num1 não é o maior número')
```
